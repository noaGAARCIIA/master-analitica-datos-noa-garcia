## Imports

```python
import os
import pandas as pd
import numpy as np
from datetime import datetime
```

## Parametros

```python
silver_path = os.path.join(
    os.curdir, "Data_lake", "Silver"
)
gold_path = os.path.join(
    os.curdir, "Data_lake", "Gold"
)


silver_file_name = "marketing_campaing_silver.parquet"
gold_file_name = "marketing_campaing_gold.parquet"



silver_file_path = os.path.join(
    silver_path, silver_file_name
)
gold_file_path = os.path.join(
    gold_path, gold_file_name
)
```

## Main

```python
silver_marketing_campaign_df = pd.read_parquet(silver_file_path) #Se carga archivo silver
```

```python
silver_marketing_campaign_df.head() #Se comprueba que archivo bronze ha cargado correctamente
```
