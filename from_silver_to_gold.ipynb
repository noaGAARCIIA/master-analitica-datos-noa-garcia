## Imports

```python
import os
import pandas as pd
import numpy as np
from datetime import datetime
```

## Parametros

```python
silver_path = os.path.join(
    os.curdir, "Data_lake", "Silver"
)
gold_path = os.path.join(
    os.curdir, "Data_lake", "Gold"
)


silver_file_name = "marketing_campaing_silver.parquet"
gold_file_name = "marketing_campaing_gold.parquet"



silver_file_path = os.path.join(
    silver_path, silver_file_name
)
gold_file_path = os.path.join(
    gold_path, gold_file_name
)
```

## Main

```python
silver_marketing_campaign_df = pd.read_parquet(silver_file_path) #Se carga archivo silver
```

```python
silver_marketing_campaign_df.head() #Se comprueba que archivo bronze ha cargado correctamente
```

## Eliminar nulos

```python
#Eliminamos los valores NaN que quedan en el df para que no den problemas en las correlaciones y nos graficos
```

```python
#Código que nos muestra en qué columnas tenemos nulos:
columnas_con_nan = silver_marketing_campaign_df.columns[silver_marketing_campaign_df.isna().any()].tolist()
print("Columnas con valores NaN:", columnas_con_nan)
```

```python
#Eliminamos valores NaN
silver_marketing_campaign_sin_nulos_df = silver_marketing_campaign_df
silver_marketing_campaign_sin_nulos_df = silver_marketing_campaign_sin_nulos_df.dropna(subset=["Conversion_Rate", "Clicks", "Engagement_Score"])
```

```python
#comprobamos que se han eliminado todos los NaN
columnas_con_nan_comprobacion = silver_marketing_campaign_sin_nulos_df.columns[silver_marketing_campaign_sin_nulos_df.isna().any()].tolist()
print("Columnas con valores NaN:", columnas_con_nan_comprobacion)
```

```python
#"Clicks" es tipo float y ahora que podemos, porque ya no hay valores NaN, lo pasamos a tipo int
silver_marketing_campaign_sin_nulos_df["Clicks"].astype(int)
```

## Guardar

```python
# Crear carpeta
os.makedirs(gold_path, exist_ok=True)
```

```python
#Añadir el timestamp de carga
silver_mc_times_tamp_df = silver_marketing_campaign_sin_nulos_df

silver_mc_times_tamp_df['_GoldTimestamp'] = datetime.now()
```
