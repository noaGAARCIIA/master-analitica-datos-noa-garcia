## Imports

```python
import os
import pandas as pd
from datetime import datetime
```

## Parametros

```python
source_file_path = os.path.join(os.curdir, "marketing_campaign_dataset_noa.csv")
```

```python
bronze_file_name = "marketing_campaign_bronze.parquet"
bronze_path = os.path.join(os.curdir, "Data_lake", "Bronze")
bronze_file_path = os.path.join(bronze_path, bronze_file_name)
```

```python
os.makedirs(bronze_path, exist_ok=True) #Se crean carpetas indicadas en bronze_path
```

## Main

```python
marketing_campaign_df = pd.read_csv(source_file_path) #Se carga el csv a un DataFrame
```

```python
marketing_campaign_df.head() #Se comprueba queel csv esté correctamente cragado
```

```python
marketing_campaign_df["_BronzeTimestamp"] = datetime.now() #Se añade timestamp de carga
```

```python
marketing_campaign_df.head() #Se comprueba que se ha añadido correctamente el timestamp
```

```python
marketing_campaign_df.to_parquet(bronze_file_path, index=False)
```
